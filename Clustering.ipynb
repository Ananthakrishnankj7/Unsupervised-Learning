# K-Means Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mtp
df = pd.read_csv("Mall_Customers.csv")
df.head()
x=df.iloc[:,[3,4]].values
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5, random_state=0, init="k-means++")
y_predict=kmeans.fit_predict(x)
k=['blue','green','red','cyan','magenta']
l=['cluster 1','cluster 2','cluster 3','cluster 4','cluster 5']
for i in range(5):
    mtp.scatter(x[y_predict == i, 0],x[y_predict == i, 1], s=100, c=k[i], label=l[i])
mtp.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', label = 'Centroid')   
mtp.title('Clusters of customers')  
mtp.xlabel('Annual Income (k$)')  
mtp.ylabel('Spending Score (1-100)')  
mtp.legend()  
mtp.show()  
```

# HIERARCHICAL CLUSTERING

```python
import scipy.cluster.hierarchy as shc  
dendro = shc.dendrogram(shc.linkage(x, method="ward"))  
mtp.title("Dendrogram Plot")  
mtp.ylabel("Euclidean Distances")  
mtp.xlabel("Customers")  
mtp.show()  
from sklearn.cluster import AgglomerativeClustering  
hc= AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')  
y_pred= hc.fit_predict(x)  
k=['blue','green','red','cyan','magenta']
l=['cluster 1','cluster 2','cluster 3','cluster 4','cluster 5']
for i in range(5):
    mtp.scatter(x[y_pred == i, 0],x[y_pred == i, 1], s=100, c=k[i], label=l[i])   
mtp.title('Clusters of customers')  
mtp.xlabel('Annual Income (k$)')  
mtp.ylabel('Spending Score (1-100)')  
mtp.legend()  
mtp.show()  
```
